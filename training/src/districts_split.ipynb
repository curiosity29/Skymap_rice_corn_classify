```python
import math
def generate_grid(bbox, size):
    x = math.ceil((bbox[2] - bbox[0]) / size)
    y = math.ceil((bbox[3] - bbox[1]) / size)
    x, y 
    box_grids = []
    for i in range(x):
        for j in range(y):
            box_grids.append([bbox[0]+ size*i, bbox[1]+ size*j, bbox[0]+ size*(i+1), bbox[1]+ size*(j+1)])
    return box_grids
```

```python
def drop_duplicates(df):
    df = dgpd.from_geopandas(df, npartitions=100)
    result = df.drop_duplicates(subset=['FID'], ignore_index=True).compute()
    return result
```

```python
def clean_farms(_farms, bbox):
    print('Clipping farm ...')
    land = gpd.read_file("../landuse/final_te_cropland_new.fgb", bbox=tuple(bbox), engine='pyogrio')
    land = dgpd.from_geopandas(land, npartitions=100)
    land = land.clip(mask=tuple(bbox)).compute()
    land = land.loc[land.geometry.is_valid]
    print('Landuse shape', land.shape)
    farms = _farms.clip(mask=land).compute()
    print('Farm shape after clipping', farms.shape)
    farms = dgpd.from_geopandas(farms, npartitions=100)

    print('Cleaning farm ...')
    farms = farms.assign(geometry=farms.buffer(3).buffer(-3))
    farms = farms.assign(shape_index=(4*np.pi*farms.area / farms.length**2)).assign(convex_index=farms.area / farms.convex_hull.area)
    farms = farms.assign(logic = ((farms.shape_index > 0.3) | (farms.convex_index > 0.8)) & (farms.area > 100) & (farms.area < 50000))
    farms = farms[farms.logic]
    farms = farms.compute()
    print('Final farm shape:', farms.shape)

    return farms
```
